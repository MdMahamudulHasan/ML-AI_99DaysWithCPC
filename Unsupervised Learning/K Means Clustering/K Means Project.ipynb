## K Means Cluster Creation
Now it's time to create the Cluster labels.
#### Import KMeans from Scikit Learn.

```python
from sklearn.cluster import KMeans
```

#### Create an instance of a K Means model with 2 clusters.

```python
kmeans = KMeans(n_clusters=2, n_init=100)
```

#### Fit the model to all the data except for the Private label.

```python
kmeans.fit(df.drop('Private', axis=1))
```

#### What are the cluster center vectors?

```python
kmeans.cluster_centers_
```

### Evaluation
There is no perfect way to evaluate clustering if you don't have the labels, however since this is just an exercise. we do have the labels, so we take advantages of this to evaluate our clusters, keep in mind , you usually won't have this luxury in the real world.
#### Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.

```python
def converter(private):
    if private == 'Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
df.head()
```

#### Create a confusion matrix and classification report to see how well the Kmeans clustering worked without given any labels.

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(df['Cluster'], kmeans.labels_))
```

```python
print(classification_report(df['Cluster'], kmeans.labels_))
```
