#### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# set a nice style for the visualizations
sns.set(style='whitegrid')
```

#### Define Pre-Created Data

```python
# Define vectors
vector1 = np.array([2, 3, 5])
vector2 = np.array([4, 1, 7])

# Define matrices
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

print("Vector 1:", vector1)
print("Vector 2:", vector2)
print("\nMatrix 1:\n", matrix1)
print("Matrix 2:\n", matrix2)
```

### Perform Operations on Vectors and Visualize

#### Vector addition and subtraction

```python
vector_addition = vector1 + vector2
vector_subtraction = vector1 - vector2

print("Vector Addition:", vector_addition)
print("Vector Subtraction:", vector_subtraction)
```

#### Simple bar plots to show vector addition

```python
x = np.arange(len(vector1)) 

plt.figure(figsize=(10, 4))
plt.bar(x, vector_addition, color='skyblue')
plt.xticks(x, [f"Element {i+1}" for i in x])
plt.title("Vector Addition")

plt.show()
```

#### Simple bar plots to show vector subtraction

```python
plt.figure(figsize=(10, 4))
plt.bar(x, vector_subtraction, color='salmon')
plt.xticks(x, [f"Element {i+1}" for i in x])
plt.title("Vector Subtraction")

plt.show()
```

#### Scalar Multiplication and Visualization

```python
scalar = 2
scaled_vector = scalar * vector1
print("Scaled Vector (Vector 1 * 2):", scaled_vector)

# Plot scalar multiplication result
plt.figure(figsize=(6, 4))
plt.bar(x, scaled_vector, color='lightgreen')
plt.xticks(x, [f"Element {i+1}" for i in x])
plt.title("Scalar Multiplication of Vector 1 by 2")
plt.show()
```

#### Dot Product

```python
dot_product = np.dot(vector1, vector2)
print("Dot Product of Vector 1 and Vector 2:", dot_product)
```

### Perform Operations on Matrix and Visualize

#### Matrix Addition and Subtraction

```python
matrix_addition = matrix1 + matrix2
matrix_subtraction = matrix1 - matrix2

print("Matrix Addition:\n", matrix_addition)
print("Matrix Subtraction:\n", matrix_subtraction)
```

#### Simple bar plots to show Matrix addition

```python
plt.figure(figsize=(10, 4))
sns.heatmap(matrix_addition, annot=True, cmap='Blues', cbar=False)
plt.title("Matrix Addition")
plt.show()
```

#### Simple bar plots to show Matrix subtraction

```python
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])
matrix_subtraction = matrix1 - matrix2

print("Matrix 1:\n", matrix1)
print("Matrix 2:\n", matrix2)
print("Matrix Subtraction:\n", matrix_subtraction)

plt.figure(figsize=(6, 4))
sns.heatmap(matrix_subtraction, annot=True, fmt=".1f", cmap='Blues', cbar=False, annot_kws={"size": 10})
plt.title("Matrix Subtraction")
plt.tight_layout()
plt.show()
```

#### Scalar Multiplication

```python
# Scalar multiplication on a matrix
scalar_matrix = 3 * matrix1
print("Matrix 1 * 3:\n", scalar_matrix)

# Heatmap for scalar multiplication
plt.figure(figsize=(6, 4))
sns.heatmap(scalar_matrix, annot=True, cmap='Greens', cbar=False)
plt.title("Scalar Multiplication of Matrix 1 by 3")
plt.show()
```

### Matrix Multiplication

```python
# Matrix multiplication
matrix_multiplication = np.dot(matrix1, matrix2)
print("Matrix Multiplication (Matrix 1 * Matrix 2):\n", matrix_multiplication)

# Heatmap for matrix multiplication
plt.figure(figsize=(6, 4))
sns.heatmap(matrix_multiplication, annot=True, cmap='Purples', cbar=False)
plt.title("Matrix Multiplication Result")
plt.show()
```

#### Transpose of a Matrix

```python
# Transpose of a matrix
matrix1_transpose = matrix1.T
print("Transpose of Matrix 1:\n", matrix1_transpose)

# Heatmap for transpose
plt.figure(figsize=(6, 4))
sns.heatmap(matrix1_transpose, annot=True, cmap='Oranges', cbar=False)
plt.title("Transpose of Matrix 1")
plt.show()
```
