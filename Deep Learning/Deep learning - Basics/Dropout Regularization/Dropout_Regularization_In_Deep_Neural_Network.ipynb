```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/sonar_dataset.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe().transpose()
```

```python
data.columns
```

```python
data['R'].value_counts()
```

```python
y = pd.get_dummies(data['R'], drop_first=True)
```

```python
y
```

```python
y.value_counts()
```

```python
X = data.drop('R', axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

## **Model without Dropout Layer**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
Early_stop = EarlyStopping(monitor='val_loss', verbose=1, patience=25)
```

```python
model = Sequential()

model.add(Dense(60, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), epochs=100, callbacks=[Early_stop], batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = np.round(y_pred)
```

```python
y_pred[:5]
```

```python
y_test[:5]
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot=True)
```

## **Model with Dropout Layer**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model1 = Sequential()

model1.add(Dense(60, activation='relu'))
model1.add(Dropout(0.3))
model1.add(Dense(30, activation='relu'))
model1.add(Dropout(0.3))
model1.add(Dense(15, activation='relu'))
model1.add(Dropout(0.3))
model1.add(Dense(1, activation='sigmoid'))

model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model1.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), epochs=100, callbacks=[Early_stop], batch_size=8)
```

```python
model1.evaluate(X_test, y_test)
```

```python
y_pred = model1.predict(X_test)
```

```python
y_pred = np.round(y_pred)
```

```python
y_pred[:5]
```

```python
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot=True)
```
