```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x**2
```

```python
x
```

```python
y
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y)

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='#FF8C00') # RGB HEX code

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', linewidth=3, alpha=0.5)

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', lw=3, alpha=0.5)

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', lw=3, linestyle='--')

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', lw=3, ls='--')

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', lw=.5, linestyle='-', marker='o',)

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', lw=.5, linestyle='-', marker='o', markersize=20)

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', lw=1, linestyle='-', marker='o', markersize=20, markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='g')

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', lw=3, ls='--')

ax.set_xlim(0,1)
ax.set_ylim(0,2)

plt.show()
```
