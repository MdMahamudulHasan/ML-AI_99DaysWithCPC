#### This is Highly imbalanced Dataset
0--> Legit Transections

1--> Fraudlent Transections

#### Separating the legit and fraudlent transactions

```python
legit = credit_card_data[credit_card_data['Class'] == 0]
fraud = credit_card_data[credit_card_data['Class'] == 1]
```

```python
legit.shape
```

```python
fraud.shape
```

#### Under-sampling method
Build a sample dataset containing similar distribution of legit and fraudlent Transections

```python
legit_sample = legit.sample(n=492)
```

```python
legit_sample.shape
```

#### Now Concatenate the Two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```
